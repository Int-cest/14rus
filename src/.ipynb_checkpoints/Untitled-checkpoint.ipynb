```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import os
```

```python
# Функция загрузки файла
def download_file(url, folder):
    response = requests.get(url, stream=True)
    filename = os.path.join(folder, os.path.basename(urlparse(url).path))
    if not filename:
        filename = os.path.join(folder, "unnamed")
    with open(filename, 'wb') as f:
        for chunk in response.iter_content(1024):
            f.write(chunk)
    print(f"Downloaded: {filename}")

# Основная функция краулера
def crawler(start_url, folder, visited=None, depth=1, save2html=True, files = ('.doc', '.docx')):
    if visited is None:
        visited = set()

    if depth == 0 or start_url in visited:
        return

    visited.add(start_url)

    try:
        response = requests.get(start_url, timeout=10)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Создание папки, если её нет
        if not os.path.exists(folder):
            os.makedirs(folder)

        # Сохранение текущей страницы HTML
        if save2html:
            page_filename = os.path.join(folder, f"page_{len(visited)}.html")
            with open(page_filename, 'w', encoding='utf-8') as file:
                file.write(response.text)
            print(f"Saved page: {page_filename}")

        # Сбор ссылок на картинки и pdf
        for tag, attr in [('img', 'src'), ('a', 'href')]:
            for link in soup.find_all(tag):
                file_url = urljoin(start_url, link.get(attr)) #
                if file_url.endswith(files):
                    download_file(file_url, folder)

        # Переход по всем ссылкам (включая внешние)
        for a_tag in soup.find_all('a', href=True):
            href = urljoin(start_url, a_tag['href'])
            if href.startswith('http'):
                crawler(href, folder, visited, depth - 1, save2html , files)

    except Exception as e:
        print(f"Error crawling {start_url}: {e}")
```

```python
start_url = "https://vuzopedia.ru/vuz"
save_folder = "dataset3"
max_depth = 6
# '.png', '.jpg', '.jpeg', '.gif','.pdf', '.xls', '.doc', '.docx', '.rtf', '.txt', '.csv', '.json'
crawler(start_url, save_folder, depth=max_depth, save2html = False, files=('.pdf', '.xls', '.doc', '.docx', '.rtf', '.txt', '.csv', '.json'))
```

```python
from parcer import *
from config import PATH_DATA
import csv
```

```python

factory = ParserFactory()
print(f"--- Начинаю сканирование директории: {PATH_DATA} ---")
results = factory.scan_directory(PATH_DATA)
print(f"--- Обработка завершена. Файлов: {len(results)} ---")

output_file_csv = "result.csv"

try:
    with open(output_file_csv, "w", encoding="utf-8", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=["path", "content"], extrasaction='ignore')
        
        writer.writeheader()
        for res in results:
            if res.get("content"):
                res["content"] = " ".join(res["content"].split())
            writer.writerow(res)
            
    print(f"Результаты парсинга успешно сохранены в {output_file_csv}")

except Exception as e:
    print(f"Ошибка при сохранении CSV: {e}")
```
